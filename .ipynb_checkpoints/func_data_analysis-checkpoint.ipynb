### Structural MRI data inspection

```python
# Load libraries

import nibabel as nib
from nilearn import plotting, image
import numpy as np
```

```python
# Loading MRI image

mri_path = 'data/sub-01_space-MNI152NLin2009cAsym_desc-preproc_T1w.nii.gz'
mri = image.load_img(mri_path)
```

```python
# Print dimensions of MRI image
print(mri.shape)

# Print type of MRI image
print(type(mri))
```

```python
# Print MRI image in Nilearn (use plotting.plot_anat)

plotting.plot_anat(mri)
```

```python
# Make an interactive plotting (use plotting.view; bg_img=None)

plotting.view_img(mri, bg_img=None)
```

```python
# Convert nibabel file to numpy array
mri_array = mri.get_fdata()

# Check and print type and dimensions
print(type(mri), mri.shape)
print(type(mri_array), mri_array.shape)
```

```python
# Plot one slice of the image (HINT: try to use np.rot90())
axial = np.rot90(mri_array[:,:,90])
coronal = np.rot90(mri_array[:,90,:])

plt.imshow(axial)
plt.imshow(coronal)
```
